```python
sorted_data = sorted(df['Site Energy Use (GJ)'])

# Calculate Q1, Q3
q1 = np.percentile(sorted_data, 25)
q3 = np.percentile(sorted_data, 75)

# Calculate IQR
iqr = q3 - q1

# Calculate lower and upper bounds
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Find outliers
outliers = [x for x in sorted_data if x < lower_bound or x > upper_bound]

num_outliers = len(outliers)

# Calculate the ratio of outliers to the total data
ratio_outliers_to_total = num_outliers / len(df)

print("Ratio of outliers to total data:", ratio_outliers_to_total)
print("Number of outliers:", num_outliers)
print("Outliers:", outliers)
```

```python
from scipy import stats

mean = np.mean(df['Site Energy Use (GJ)'])

# Calculate median
median = np.median(df['Site Energy Use (GJ)'])

# Calculate mode
mode = stats.mode(df['Site Energy Use (GJ)'])

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
def remove_outliers(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_filtered

# Remove outliers from column 'B'
df_cleaned = remove_outliers(df,'Site Energy Use (GJ)')

# Calculate mean, median, and mode of column 'B' after removing outliers
mean_B = df_cleaned['Site Energy Use (GJ)'].mean()
median_B = df_cleaned['Site Energy Use (GJ)'].median()
mode_B = stats.mode(df_cleaned['Site Energy Use (GJ)'])

print("Mean of column B after removing outliers:", mean_B)
print("Median of column B after removing outliers:", median_B)
print("Mode of column B after removing outliers:", mode_B)
```
